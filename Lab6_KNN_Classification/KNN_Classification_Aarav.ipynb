```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
data = load_wine()
```

```python
X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# PART A: BEFORE PCA

print("----- BEFORE PCA -----")

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train_scaled, y_train)

y_pred_before = knn.predict(X_test_scaled)
acc_before = accuracy_score(y_test, y_pred_before)

print("Accuracy BEFORE PCA:", acc_before)
```

```python
# PART B: APPLY PCA

pca = PCA(n_components=2)

X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
# PART C: AFTER PCA

print("\n----- AFTER PCA -----")

knn_pca = KNeighborsClassifier(n_neighbors=5)

knn_pca.fit(X_train_pca, y_train)

y_pred_after = knn_pca.predict(X_test_pca)
acc_after = accuracy_score(y_test, y_pred_after)

print("Accuracy AFTER PCA:", acc_after)
```

```python
# BONUS: Explained Variance

print("\nExplained Variance Ratio:", pca.explained_variance_ratio_)
```

# INTERPRETATION:
# Before applying PCA, the model used all original features and achieved high accuracy.
# After applying PCA, the dimensionality was reduced to 2 principal components.
# Despite reducing features, the accuracy improved, indicating that PCA removed noise and redundant information.
# PCA helped in improving model performance while simplifying the dataset.
# The explained variance ratio shows that the selected components capture a significant portion of the data variance.
