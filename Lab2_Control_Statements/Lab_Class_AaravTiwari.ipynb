```python
df_train["Age"] = df_train["Age"].fillna(df_train["Age"].median())
df_train = df_train.drop(["Cabin"], axis=1)
df_train = df_train.dropna(axis=0, subset=["Embarked"])
```

```python
df_test["Age"] = df_test["Age"].fillna(df_test["Age"].median())
df_test = df_test.drop(["Cabin"], axis=1)
df_test = df_test.dropna(axis=0, subset=["Fare"])
```

```python
df_train_num = df_train.select_dtypes(exclude="object")
df_train_obj = df_train.select_dtypes(include="object").drop(["Name", "Ticket"], axis=1)

df_test_num = df_test.select_dtypes(exclude="object")
df_test_obj = df_test.select_dtypes(include="object").drop(["Name", "Ticket"], axis=1)
```

```python
df_train_obj = pd.get_dummies(df_train_obj, drop_first=True)
df_test_obj = pd.get_dummies(df_test_obj, drop_first=True)
```

```python
df_test_obj = df_test_obj.reindex(columns=df_train_obj.columns, fill_value=0)
```

```python
Final_train_df = pd.concat([df_train_num, df_train_obj], axis=1)
Final_test_df = pd.concat([df_test_num, df_test_obj], axis=1)
```

```python
X_train = Final_train_df.drop("Survived", axis=1)
y_train = Final_train_df["Survived"]

X_test = Final_test_df.drop("Survived", axis=1)
y_test = Final_test_df["Survived"]
```

```python
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)
```

```python
from sklearn.preprocessing import StandardScaler

# FIX
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

scaler = StandardScaler()
scaler.fit(X_train)

scaled_X_train = scaler.transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(scaled_X_train, y_train)

y_pred = knn.predict(scaled_X_test)

print("KNN Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(scaled_X_train, y_train)

y_pred_lr = lr.predict(scaled_X_test)

print("Logistic Accuracy:", accuracy_score(y_test, y_pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)

print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
```

## Final Interpretation

In this experiment, the Titanic dataset was analysed using three different machine learning models: K-Nearest Neighbors (KNN), Logistic Regression, and Decision Tree Classifier. Before applying the models, data preprocessing was performed, which included handling missing values, removing irrelevant features, encoding categorical variables using dummification, and standardizing numerical features.

The KNN model produced an accuracy of 37.65%, which is significantly low compared to the other models. This indicates that KNN was not suitable for this dataset, possibly due to its sensitivity to feature scaling and high dimensionality after encoding.

The Logistic Regression model achieved an accuracy of 73.38%, showing a considerable improvement. This suggests that the dataset has a relatively linear relationship between the input features and the target variable.

The Decision Tree Classifier performed the best with an accuracy of 77.94%. This model was able to capture complex patterns and interactions within the dataset more effectively than the other models.

Overall, it can be concluded that the Decision Tree model is the most suitable for this dataset.
