```python
# Problem 1: Confusion Matrix Metrics

# Given values
TP = 50
FP = 8
TN = 32
FN = 10

# Total predictions
total = TP + FP + TN + FN
print("Total Predictions:", total)

# Accuracy
accuracy = (TP + TN) / total
print("Accuracy:", accuracy)

# Precision
precision = TP / (TP + FP)
print("Precision:", precision)

# Recall
recall = TP / (TP + FN)
print("Recall:", recall)

# F1 Score
f1_score = 2 * precision * recall / (precision + recall)
print("F1 Score:", f1_score)
```

```python
# Problem 2: R² (Coefficient of Determination)

# Given values
SSR = 420   # Sum of Squares due to Regression
SSE = 180   # Sum of Squares due to Error

# Total Sum of Squares
SST = SSR + SSE
print("Total Sum of Squares (SST):", SST)

# R-squared calculation
R_squared = SSR / SST
print("R-squared:", R_squared)

# Interpretation
print("Model explains more than 70% variance:", R_squared > 0.7)
```

```python
# Problem 4: Cook’s Distance

# Given values
ei_squared = 9
MSE = 3
p = 3
h_ii = 0.2
n = 25

# Cook's Distance calculation
cooks_distance = (ei_squared / (p * MSE)) * (h_ii / ((1 - h_ii) ** 2))
print("Cook's Distance:", cooks_distance)

# Threshold
threshold = 4 / n
print("Threshold (4/n):", threshold)

# Decision
print("Is the observation influential?", cooks_distance > threshold)
```

```python
# Problem 5: Simple Linear Regression Prediction

# Given values
x = 12
slope = 2.5
intercept = 10

# Prediction
y = slope * x + intercept
print("Predicted value of y:", y)

# Explanation
print("Slope (2.5): Change in y for one unit change in x")
print("Intercept (10): Value of y when x = 0")
```
